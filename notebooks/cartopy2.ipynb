```python
import matplotlib.pyplot as plt
%matplotlib inline

import cartopy.crs as ccrs
import cartopy.feature as cfeat
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

# Region extents
EXTENTS = {
    'tropical_north_atlantic': [-105, -10, 0, 50],
    'medium': [-90, -20, 0, 27],
    'caribbean': [-65, -53, 10, 19],
    'lesser_antilles': [-63.0, -59.0, 10.0, 17.5],
    'grenada': [-62.3, -61.0, 11.7, 12.9],
}

# Style
STYLE = {
    'bg': '#0a1628',
    'ocean': '#1a3a5c',
    'land': '#2d4a3e',
    'coast': '#88b4d4',
    'grid': '#ffffff',
    'text': '#e0e0e0',
}
```

```python
def atlantic_map(region="grenada", figsize=(12, 10)):
    fig = plt.figure(figsize=figsize, facecolor=STYLE["bg"])
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
    ax.set_facecolor(STYLE["ocean"])

    ax.add_feature(cfeat.LAND, facecolor=STYLE["land"])
    ax.add_feature(cfeat.OCEAN, facecolor=STYLE["ocean"])
    ax.add_feature(cfeat.COASTLINE, linewidth=0.8, edgecolor=STYLE["coast"])
    ax.add_feature(cfeat.BORDERS, linestyle=":", linewidth=0.5)
    ax.add_feature(cfeat.LAKES, alpha=0.5)
    ax.add_feature(cfeat.RIVERS, linewidth=0.5)

    gl = ax.gridlines(
        draw_labels=True, linewidth=0.5, color=STYLE["grid"], alpha=0.3, linestyle="--"
    )
    gl.top_labels = gl.right_labels = False
    gl.xformatter, gl.yformatter = LONGITUDE_FORMATTER, LATITUDE_FORMATTER
    gl.xlabel_style = gl.ylabel_style = {"size": 9, "color": STYLE["text"]}

    ax.set_extent(EXTENTS[region], crs=ccrs.PlateCarree())
    ax.set_title(
        region.replace("_", " ").title(), fontsize=14, color=STYLE["text"], pad=12
    )

    return fig, ax
```

```python
for region in EXTENTS:
    atlantic_map(region)
    plt.show()
```
