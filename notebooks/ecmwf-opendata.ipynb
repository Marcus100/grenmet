```python
# ECMWF Open Data Downloader
# Modular notebook for downloading ECMWF forecast data

from ecmwf.opendata import Client
from pathlib import Path
from datetime import datetime, timezone

utc_now = datetime.now(timezone.utc)

# MODEL & PATHS
MODEL = "ifs"  # "ifs", "aifs-single", "aifs-ens"
DATA_DIR = Path("data/ecmwf") / MODEL / "raw"

# Run selection (None = latest available)
DATE = None  # e.g., "2025-01-03" or None
TIME = None  # 0, 6, 12, 18 or None



# CLIENT
client = Client(source="ecmwf", model=MODEL)

print(f"Model: {MODEL}")
print(f"Data dir: {DATA_DIR}")
print(utc_now)
print(client.latest())
```

```python
client.retrieve(
    step=240,
    type="fc",
    param="msl",
    target="data.grib2",
)
```

```python
client = Client(
    source="ecmwf",
    model=MODEL,
    resol="0p25",
    preserve_request_order=False,
    infer_stream_keyword=True,
)
type(client)
print(client)
```

```python
# client = Client(source="ecmwf")

result = client.retrieve(
    time=0,
    type="fc",
    step=24,
    param=["2t", "msl"],
    target=f"{DATA_DIR}/data.grib2",
)
```
